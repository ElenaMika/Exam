```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount = True) 
togliatti = pd.read_csv('/content/drive/MyDrive/Питон/data_26_01_2023.csv', delimiter = ',', encoding= 'utf-8')
togliatti.tail(50)
```

Очистим данные

```python
# данные, разделенные запятой в столбце "Район", сохраним в разные столбцы ('Район_' и 'Квартал') и удалим исходный столбец
togliatti_new = togliatti['Район'].str.split(',',expand=True)
togliatti_new.columns=['Район_','Квартал']
for column in togliatti_new.columns:
    togliatti_new[column] = togliatti_new[column].str.replace(column+',','')
togliatti_final = pd.concat([togliatti,togliatti_new],axis=1).drop('Район',axis=1)

# удалим нечисловые значения из столбца 'Цена' и удалим строки, которые не содержат чисел в столбце 'Цена'
for column in togliatti_final.columns:
    togliatti_final['Цена'] = togliatti_final['Цена'].str.extract('([0-9]+)')
togliatti_final = togliatti_final.dropna()
togliatti_final = togliatti_final.drop_duplicates()

# удалим нечисловые значения из столбца 'Жилая площадь' и удалим строки, которые не содержат чисел в столбце 'Жилая площадь'
for column in togliatti_final.columns:
    togliatti_final['Жилая площадь'] = togliatti_final['Жилая площадь'].astype(str).str.extract('([0-9]+)')

# удалим из датафрейма районы, которые не относятся к г.о.Тольятти
togliatti_final = togliatti_final.loc[togliatti_final['Район_'] != 'Жигулевск']
togliatti_final = togliatti_final.loc[togliatti_final['Район_'] != 'Шигонский']
togliatti_final = togliatti_final.loc[togliatti_final['Район_'] != 'Сызранский']
togliatti_final = togliatti_final.loc[togliatti_final['Район_'] != 'Ставропольский']

# оставим только числа в столбце "Квартал"
for column in togliatti_final.columns:
    togliatti_final['Квартал'] = togliatti_final['Квартал'].str.extract('([0-9]+)')

# данные, разделенные слешем в столбце "Этажность", сохраним в разные столбцы ('Этаж' и 'Всего этажей') и удалим исходный столбец
togliatti_final[['Этаж','Всего этажей']] = togliatti_final['Этажность'].str.split('/', 1, expand=True)

togliatti_final1 = togliatti_final.drop('Этажность',axis=1)

# данные, разделенные слешем в столбце "Планировка", сохраним в разные столбцы ('Планировка_' и 'Материал стен') и удалим исходный столбец
togliatti_final1[['Планировка_','Материал стен']] = togliatti_final1['Планировка'].str.split('/', 1, expand=True)

togliatti_final2 = togliatti_final1.drop('Планировка',axis=1)

togliatti_final2.head(100)
```

```python
togliatti_final2 = togliatti_final2.rename(columns={"Планировка_": "plan"}) # временно поменяем название столбца "Планировка_" на англоязычное
marks_list1 = togliatti_final2.plan.unique() # посмотрим, какие есть варианты планировки.
marks_list1.sort()
print(marks_list1)    # выведем список всех уникальных значений
mapping = []
for i in range(1, (len(marks_list1)+1)):   # создадим список состоящий из чисел от 1 до количества уникальных значений в столбце "plan"
  mapping.append(i)
print(mapping)
```

```python
myDict = {marks_list1[i]: mapping[i] for i in range(len(marks_list1))}    # создадим словарь, в котором каждому значению присвоим число
print(myDict)
```
